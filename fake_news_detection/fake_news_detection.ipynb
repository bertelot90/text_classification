# 3. Prétraitement de texte

Pour le prétraitement, nous allons effectuer les tâches suivantes:
- Tockenisation
- Suppression des stopwords
- Lematisation
- Racinisation

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.snowball import EnglishStemmer
import string
import numpy as np
```

### 3.1.  Tokenisation

Nous allons ici découper les différents tweets et tags en tockens (mots)

```python
def tokenisation(tweet):
    
    if tweet is not np.nan:
        try:
            tokens = [word.replace('\n', '').strip() for word in word_tokenize(tweet)]
        except:
            print(type(tweet))
            print(type(np.nan))
        if len(tokens) > 0:
            return tokens
        return None
    else:
        return None
```

```python
data['tockenized_text'] = data['Text'].apply(tokenisation)
data['tockenized_text_tag'] = data['Text_Tag'].apply(tokenisation)
```

```python
data.head()
```

### 3.2. Suppression des mots vides (Stopwords)

Nous allons supprimer les stopwords sur les tokens créé précédemment

```python
# Liste des stopwords
print(stopwords.words('english'))
```

```python
# Liste des ponctuations
punctuation = list(string.punctuation) # une liste avec ponctuations

np.array(punctuation)
```

```python
def delete_stopwords(tweet):
    
    if tweet is not None:
        tokens = [word for word in tweet if word.lower() not in stopwords.words('english') + punctuation]
        tokens = [token for token in tokens if len(token) > 1]
        if len(tokens) > 0:
            return tokens
        return None
```

```python
data['text_without_stopword'] = data['tockenized_text'].apply(delete_stopwords)
data['text_tag_without_stopword'] = data['tockenized_text_tag'].apply(delete_stopwords)
```

```python
data.head()
```

### 3.3. Lematisation

Nous allons appliquer la lematisation sur le données sur lesquels nous avons enlevé les stopwords

```python
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
correspondaces = {'ADJ': "a", 'ADJ_SAT': "s", 'ADV': "r", 'NN': "n", 'VB': "v", "NNS": "n", "NNP": "n", "VBD": "v",
                 "VBN": "v", "VBZ": "v", "VBG": "v"}
```

```python
def lemmatisation(tweet):
    
    if tweet is not None:
        #liste_mots = word_tokenize(phrase)
        tokens = [0] * len(tweet)
        for ind, mot in enumerate(tweet):
            a, b = nltk.pos_tag([mot])[0]

            if b in list(correspondaces.keys()):
                tokens[ind] = lemmatizer.lemmatize(mot, pos=correspondaces[b])
            else:
                tokens[ind] = mot

        tokens = [token for token in tokens if len(token) > 1]

        if len(tokens) > 0:
            return tokens
        return None
```

```python
data['text_cleaned_stopword_and_lemmatisation'] = data['text_without_stopword'].apply(lemmatisation)
data['text_tag_cleaned_stopword_and_lemmatisation'] = data['text_tag_without_stopword'].apply(lemmatisation)
```

```python
data.head()
```

```python
# On vérifie qu'il nya pas de tweet vide
data['text_cleaned_stopword_and_lemmatisation'].isnull().sum()
```

### 3.4. Racinisation

Nous allons appliquer la racinisation sur le données sur lesquels nous avons enlevé les stopwords

```python
stemmer = EnglishStemmer()
```

```python
def stemming(tweet):
    
    if tweet is not None:
        tokens = [stemmer.stem(word) for word in tweet]
        tokens = [token for token in tokens if len(token) > 1]

        if len(tokens) > 0:
            return tokens
        return None
```

```python
data['text_stopword_and_stemming'] = data['text_without_stopword'].apply(stemming)
data['text_tag_stopword_and_stemming'] = data['text_tag_without_stopword'].apply(stemming)
```

# 3. Découpage des données

Nous effectuons le découpage des données afin de les utiliser pour la formation et le test des modèles de machine learning.

Nous allons utiliser les données issues de la lematisation et ceux issues de la racinisation pour chaque modèle que nous allons utiliser.

Nous allons diviser les données de la manière suivante:
- 30 % pour le test
- 70 % pour le train

Nous allons fusionner les tweets et les tweets tags pour la formation des modèles

```python
text_lem = [' '.join(text) for text in data['text_cleaned_stopword_and_lemmatisation']]
text_rac = [' '.join(text) for text in data['text_stopword_and_stemming']]

text_tag_lem = []
for text in data['text_tag_cleaned_stopword_and_lemmatisation']:
    if text is not None:
        text_tag_lem.append(' '.join(text))
    else:
        text_tag_lem.append('')
        
text_tag_rac = []
for text in data['text_tag_stopword_and_stemming']:
    if text is not None:
        text_tag_rac.append(' '.join(text))
    else:
        text_tag_rac.append('')

X_train_lem, X_test_lem, y_train_lem, y_test_lem = train_test_split([tweet + tag for (tweet, tag) in zip(text_lem, text_tag_lem)], data['Labels'], test_size=0.30)
X_train_rac, X_test_rac, y_train_rac, y_test_rac = train_test_split([tweet + tag for (tweet, tag) in zip(text_rac, text_tag_rac)], data['Labels'], test_size=0.30)

print(len(X_train_lem), len(X_test_lem))
print(len(X_train_rac), len(X_test_rac))
```

```python
# Création des données textuelles dans le format correct pour le réseau neuronal
def get_train_test(train_raw_text, test_raw_text):
    
    tokenizer = text.Tokenizer(num_words=MAX_FEATURES)

    tokenizer.fit_on_texts(list(train_raw_text))
    
    train_tokenized = tokenizer.texts_to_sequences(train_raw_text)
    test_tokenized = tokenizer.texts_to_sequences(test_raw_text)
    
    return sequence.pad_sequences(train_tokenized, maxlen=VOCAB_SIZE), \
           sequence.pad_sequences(test_tokenized, maxlen=VOCAB_SIZE)
```

```python
# Création du modèle
def get_model():

    inputs = Input(shape=(VOCAB_SIZE,))
    model = Embedding(MAX_FEATURES, EMBEDDING_SIZE)(inputs)
    bidirectional_1 = Bidirectional(LSTM(64,  return_sequences=True))(model)
    bidirectional_2 = Bidirectional(LSTM(32))(bidirectional_1)
    
    model = Dense(128, activation='relu')(bidirectional_2)
    model = Dense(64, activation='relu')(model)
    dropout = Dropout(0.5)(model)

    outputs = Dense(n_out, activation="softmax")(dropout)
    model = Model(inputs=inputs, outputs=outputs)
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.summary()
    return model
```

```python
model = get_model()
```

```python
# Entrainement du modèle
def train_fit_predict(model, X_train, X_test, y):
    
    history = model.fit(X_train, y,
              batch_size=BATCH_SIZE,
              epochs=EPOCHS, verbose=1,
              validation_split=VALIDATION_SPLIT)

    return model.predict(X_test), history
```

```python
# Entrainement et prédiction
y_predicted, history = train_fit_predict(model, x_vec_train, x_vec_test, train_y_cat)
```
