# 3. Prétraitement de texte

Pour le prétraitement, nous allons effectuer les tâches suivantes:
- Suppression des liens hypertextes
- Tockenisation
- Suppression des stopwords
- Lematisation
- Racinisation

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.snowball import EnglishStemmer
import string
import numpy as np
```

### 3.1. Suppréssion des liens hypertextes

```python
import re
def remove_urls(tweet):
    tweet = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', tweet, flags=re.MULTILINE)
    return(tweet)
```

```python
data['OriginalTweet'] = data['OriginalTweet'].apply(remove_urls)
```

```python
data.head()
```

### 3.2.  Tokenisation

Nous allons ici découper les différents tweets et tags en tockens (mots)

```python
def tokenisation(tweet):
    
    if tweet is not np.nan:
        try:
            tokens = [word.replace('\n', '').strip() for word in word_tokenize(tweet)]
        except:
            print(type(tweet))
            print(type(np.nan))
        if len(tokens) > 1:
            return tokens
        return None
    else:
        return None
```

```python
data['tockenized_tweet'] = data['OriginalTweet'].apply(tokenisation)
```

```python
data.head()
```

### 3.3. Suppression des mots vides (Stopwords)

Nous allons supprimer les stopwords sur les tokens créé précédemment

```python
# Liste des stopwords
print(stopwords.words('english'))
```

```python
# Liste des ponctuations
punctuation = list(string.punctuation) # une liste avec ponctuations
punctuation.append('...')
np.array(punctuation)
```

```python
def delete_stopwords(tweet):
    
    if tweet is not None:
        tokens = [word for word in tweet if word.lower() not in stopwords.words('english') + punctuation]
        tokens = [token for token in tokens if len(token) > 1]
        if len(tokens) > 1:
            return tokens
        return None
```

```python
data['tweet_without_stopword'] = data['tockenized_tweet'].apply(delete_stopwords)
```

```python
data.head()
```

### 3.4. Lematisation

Nous allons appliquer la lematisation sur les données sur lesquels nous avons enlevé les stopwords

```python
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
correspondaces = {'ADJ': "a", 'ADJ_SAT': "s", 'ADV': "r", 'NN': "n", 'VB': "v", "NNS": "n", "NNP": "n", "VBD": "v",
                 "VBN": "v", "VBZ": "v", "VBG": "v"}
```

```python
def lemmatisation(tweet):
    
    if tweet is not None:
        tokens = [0] * len(tweet)
        for ind, mot in enumerate(tweet):
            a, b = nltk.pos_tag([mot])[0]

            if b in list(correspondaces.keys()):
                tokens[ind] = lemmatizer.lemmatize(mot, pos=correspondaces[b])
            else:
                tokens[ind] = mot

        tokens = [token for token in tokens if len(token) > 1]

        if len(tokens) > 1:
            return tokens
        return None
```

```python
data['tweet_without_stopword_lematisation'] = data['tweet_without_stopword'].apply(lemmatisation)
```

```python
data.head()
```

```python
# On vérifie qu'il nya pas de tweet vide
data['tweet_without_stopword_lematisation'].isnull().sum()
```

```python
# On supprime tous les tweet vide
data.dropna(subset=['tweet_without_stopword_lematisation'], inplace=True)
```

```python
# On vérifie qu'il nya pas de tweet vide
data['tweet_without_stopword_lematisation'].isnull().sum()
```

### 3.5. Racinisation

Nous allons appliquer la racinisation sur le données sur lesquels nous avons enlevé les stopwords

```python
stemmer = EnglishStemmer()
```

```python
def stemming(tweet):
    
    if tweet is not None:
        tokens = [stemmer.stem(word) for word in tweet]
        tokens = [token for token in tokens if len(token) > 1]

        if len(tokens) > 1:
            return tokens
        return None
```

```python
data['tweet_without_stopword_stemming'] = data['tweet_without_stopword'].apply(stemming)
```

```python
data.head()
```

```python
# On vérifie qu'il nya pas de tweet vide
data['tweet_without_stopword_stemming'].isnull().sum()
```

# 5. Neural network

```python
import ast
import os
from tensorflow.keras.layers import Dense, Embedding, Input
from tensorflow.keras.layers import Dropout, Flatten, GlobalMaxPooling1D, Bidirectional, LSTM
from tensorflow.keras.models import Model
import numpy as np
import itertools
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import (classification_report, 
                             precision_recall_fscore_support, 
                             accuracy_score)

from tensorflow.keras.preprocessing import text, sequence
from tensorflow.keras.layers import Concatenate

import tensorflow as tf
```

```python
MAX_FEATURES = 10000
VOCAB_SIZE = 1024
EMBEDDING_SIZE  = 300
BATCH_SIZE = 64
EPOCHS = 5
VALIDATION_SPLIT = 0.2
```

```python
# Création des données textuelles dans le format correct pour le réseau neuronal
def get_train_test(train_raw_text, test_raw_text):
    
    tokenizer = text.Tokenizer(num_words=MAX_FEATURES)

    tokenizer.fit_on_texts(list(train_raw_text))
    
    train_tokenized = tokenizer.texts_to_sequences(train_raw_text)
    test_tokenized = tokenizer.texts_to_sequences(test_raw_text)
    
    return sequence.pad_sequences(train_tokenized, maxlen=VOCAB_SIZE), \
           sequence.pad_sequences(test_tokenized, maxlen=VOCAB_SIZE)
```

```python
x_vec_train, x_vec_test = get_train_test(X_train_lem, X_test_lem)
len(x_vec_train), len(x_vec_test)
```

```python
# Liste des différentes classes
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
CLASSES_LIST = np.unique(data['Sentiment'])
n_out = len(CLASSES_LIST)
print(CLASSES_LIST, n_out)
le.fit(CLASSES_LIST)

y_train_encoded = le.transform(y_train_lem) 
y_test_encoded = le.transform(y_test_lem) 
train_y_cat = to_categorical(y_train_encoded, n_out)
print(y_train_encoded[:5])
```

```python
# OneHot Encoding sur les labels
train_y_cat = to_categorical(y_train_encoded, n_out)
train_y_cat
```

```python
# Création du modèle
def get_model():

    inputs = Input(shape=(VOCAB_SIZE,))
    model = Embedding(MAX_FEATURES, EMBEDDING_SIZE)(inputs)
    dropout_1 = Dropout(0.5)(model)
    bidirectional_1 = Bidirectional(LSTM(64,  return_sequences=True))(dropout_1)
    bidirectional_2 = Bidirectional(LSTM(32))(bidirectional_1)
    
    model = Dense(128, activation='relu')(bidirectional_2)
    model = Dense(64, activation='relu')(model)
    dropout_2 = Dropout(0.5)(model)

    outputs = Dense(n_out, activation="softmax")(dropout_2)
    model = Model(inputs=inputs, outputs=outputs)
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.summary()
    return model
```

```python
model = get_model()
```

```python
# Entrainement du modèle
def train_fit_predict(model, X_train, X_test, y):
    
    history = model.fit(X_train, y,
              batch_size=BATCH_SIZE,
              epochs=EPOCHS, verbose=1,
              validation_split=VALIDATION_SPLIT)

    return model.predict(X_test), history
```

```python
# Entrainement et prédiction
y_predicted, history = train_fit_predict(model, x_vec_train, x_vec_test, train_y_cat)
```

```python
y_predicted = y_predicted.argmax(1)
```

```python
def plot_graphs(history, metric):
  plt.plot(history.history[metric])
  plt.plot(history.history['val_'+metric], '')
  plt.xlabel("Epochs")
  plt.ylabel(metric)
  plt.legend([metric, 'val_'+metric])
```
